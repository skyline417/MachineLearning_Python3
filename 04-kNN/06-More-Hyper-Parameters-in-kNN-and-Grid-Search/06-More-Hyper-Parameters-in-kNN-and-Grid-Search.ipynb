## 06 网格搜索和更多kNN中的超参数

```python
import numpy as np
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=666)
```

```python
from sklearn.neighbors import KNeighborsClassifier

sk_knn_clf = KNeighborsClassifier(n_neighbors=4, weights="uniform")
sk_knn_clf.fit(X_train, y_train)
sk_knn_clf.score(X_test, y_test)
```

### Grid Search

```python
param_grid = [
    {
        'weights': ['uniform'], 
        'n_neighbors': [i for i in range(1, 11)]
    },
    {
        'weights': ['distance'],
        'n_neighbors': [i for i in range(1, 11)], 
        'p': [i for i in range(1, 6)]
    }
]
```

```python
knn_clf = KNeighborsClassifier() #对knn机器学习算法进行网格搜索 创建一个默认分类器
```

```python
from sklearn.model_selection import GridSearchCV 

grid_search = GridSearchCV(knn_clf, param_grid) #调用sklearn中的网格搜索 创建实例对象
#①对哪个分类器进行网格搜索 ②网格搜索的参数
```

```python
%%time
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

最佳分类器超参数的值和上一小节不一样（random_state都是666）

原因：评价标准不一样

网格搜索中 评价分类器准确度方式 CV：cross validation 交叉验证

比train_test_split求出的准确度更加准确

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

现在我们拿到了grid search得到的最佳参数

然后要拿这个最佳参数对应的分类器

```python
knn_clf = grid_search.best_estimator_ #把我们之前自定义的knn分类器实例 赋了这个网格搜索出来的实例
```

```python
knn_clf.predict(X_test)
```

```python
knn_clf.score(X_test, y_test)
```

n_jobs 创建很多分类器的过程可以并行处理 分配几个核 n_jobs默认是1，-1代表all

verbose输出信息 传入的整数值越大 输出的信息越多

```python
%%time
grid_search = GridSearchCV(knn_clf, param_grid, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```
