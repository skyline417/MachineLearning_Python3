### 使用scikit-learn中的kNN

```python
from sklearn.neighbors import KNeighborsClassifier #以面向对象形式包装了kNN算法
```

```python
kNN_classifier = KNeighborsClassifier(n_neighbors=6) #创建实例 传入k值
```

```python
kNN_classifier.fit(X_train, y_train) #拟合 传入训练集矩阵X 标签向量y 返回值是该对象自身
```

```python
kNN_classifier.predict(x) #报错 需要传入矩阵 每一行是要预测的样本 而x只是一个向量
```

```python
X_predict = x.reshape(1, -1) #把x reshape成矩阵 一个元素 -1：自动决定第二个维度有多少
```

```python
X_predict
```

```python
kNN_classifier.predict(X_predict)
```

```python
y_predict = kNN_classifier.predict(X_predict) #结果是一个向量 只有一个元素
```

```python
y_predict[0]
```
